```
import pandas as pd
url= 'https://raw.githubusercontent.com/karlajagg/data/master/xtrain2.csv'
data=pd.read_csv(url)
print (data)
```

```
data1 = data[['AB','COM','CHC', 'LC', 'ACC', 'CHH', 'CIM', 'ST', 'EST', 'NOMS', 'BRE']]
print(data1)
```

```
from pgmpy.models import BayesianModel

model = BayesianModel([('EST', 'AB'),('EST', 'NOMS'),('EST', 'CIM'),('CIM', 'ST'),('ST', 'COM') , ('AB','COM'),  ('NOMS','COM'),  ('NOMS','ACC'),  ('CIM','COM'),  ('CIM','NOMS'),  ('LC','NOMS'),  ('CHC','COM'),  ('CHC','AB'),  ('CHC','CHH'),  ('ACC','AB'),  ('CHH','ACC'),  ('CHH','COM'),  ('BRE','CIM'),  ('BRE','CHH')]) 
#El aprendizaje de par√°metros es la tarea de estimar los valores de las 
#distribuciones de probabilidad condicional (CPD), para las variables AB CHC CHH ACC COM NOMS TipoFALLA.
```

```
from pgmpy.estimators import ParameterEstimator
pe = ParameterEstimator(model, data)
#CONTADOR DE ESTADOS
print("\n", pe.state_counts('AB')) # conditional
print("\n", pe.state_counts('COM')) # conditional
print("\n", pe.state_counts('CHC')) # unconditional
print("\n", pe.state_counts('ACC')) # conditional
print("\n", pe.state_counts('CHH')) # unconditional
print("\n", pe.state_counts('CIM'))
print("\n", pe.state_counts('LC'))  
print("\n", pe.state_counts('ST'))
print("\n", pe.state_counts('EST')) 
print("\n", pe.state_counts('NOMS'))
print("\n", pe.state_counts('BRE')) 
```

```
#Maximum Likelihood Estimation
from pgmpy.estimators import MaximumLikelihoodEstimator
mle = MaximumLikelihoodEstimator(model, data)
print(mle.estimate_cpd('EST')) # conditional
print(mle.estimate_cpd('AB')) # conditional
print(mle.estimate_cpd('NOMS'))  # unconditional
print(mle.estimate_cpd('CIM')) # conditional
print(mle.estimate_cpd('ST')) # unconditional
print(mle.estimate_cpd('COM')) # unconditional
print(mle.estimate_cpd('ACC'))  # conditional
print(mle.estimate_cpd('LC')) # conditional
print(mle.estimate_cpd('CHC')) # unconditional
print(mle.estimate_cpd('CHH')) # unconditional
print(mle.estimate_cpd('BRE'))  # conditional
```

```
# Calibrate all CPDs of `model` using MLE:
model.fit(data, estimator=MaximumLikelihoodEstimator)
```

```
from pgmpy.estimators import BayesianEstimator
# BayesianEstimator, too, can be used via the fit()-method
# Aqui imprime todas las CPDs
model.fit(data, estimator=BayesianEstimator, prior_type="BDeu", equivalent_sample_size=10) # default equivalent_sample_size=5
for cpd in model.get_cpds():
    print(cpd)
```
